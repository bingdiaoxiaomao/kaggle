# 简介

本质上，流形学习就是个数据降维的过程。这里假设数据是一个随机样本，采样自一个高维欧式空间里的流行（manifold），流形学习的任务就是把这个高维流行映射到一个低维（例如2维）的空间里。流形学习可以分为线性算法和非线性算法，前者包括主成分分析（PCA）和线性判别分析（LDA），后者包括等距映射（Isomap），拉普拉斯特征映射（LE）等。流形学习可以用于特征的降维和提取，为后续的基于特征的分析，如聚类和分类，做铺垫，也可以直接应用于数据可视化等。

## 案例：球面映射
使用模拟数据制造一个球面，并将这个3维空间中的流形project到2维上。

```python
from time import time
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
from sklearn import manifold
from sklearn.utils import check_random_state
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
n_neighbors = 10
n_samples = 1000
random_state = check_random_state(0)
p = random_state.rand(n_samples)*(2*np.pi-0.55)
t = random_state.rand(n_samples)*np.pi
indices = ((t<((np.pi-(np.pi/8))))&(t>((np.pi/8))))
colors = p[indices]
x, y, z = np.sin(t[indices])*np.cos(p[indices]),np.sin(t[indices])*np.sin(p[indices]),np.cos(t[indices])
```

```python
fig = plt.figure(figsize=(15,8))
plt.suptitle('Manifold Learning with {} points, {} neighbors'.format(n_samples,n_neighbors),fontsize=14)
ax = fig.add_subplot(251, projection='3d')
ax.scatter(x, y, z, c=p[indices], cmap=plt.cm.rainbow)
plt.draw()
sphere_data = np.array([x,y,z]).T


# 拟合线性的流形学习模型
# LLE, LTSA, Hessian LLE和Modified LLE
methods = ['standard', 'ltsa', 'hessian', 'modified']
labels = ['LLE', 'LTSA', 'Hessian LLE', 'Modified LLE']
for i, method in enumerate(methods):
    t0 = time()
    trans_data = manifold.LocallyLinearEmbedding(n_neighbors=n_neighbors,n_components=2,method=method).fit_transform(sphere_data).T
    t1 = time()
    print('{}: {:.2f} sec'.format(methods[i],t1-t0))
    ax = fig.add_subplot(252 + i)
    plt.scatter(trans_data[0],trans_data[1],c=colors,cmap=plt.cm.rainbow)
    plt.title('{} ({:.2f} sec)'.format(labels[i],t1-t0))
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    plt.axis('tight')
    
# 拟合非线性的流行学习模型Isomap，MDS，Spectral Embedding和t-distributed stochastic neighbor embedding
# Perform Isomap Manifold learning
t0 = time()
trans_data = manifold.Isomap(n_neighbors=n_neighbors,n_components=2).fit_transform(sphere_data).T
t1 = time()
print('{}: {:.2f} sec'.format('ISO',t1-t0))
ax = fig.add_subplot(257)
plt.scatter(trans_data[0],trans_data[1],c=colors,cmap=plt.cm.rainbow)
plt.title('{} ({:.2f} sec)'.format('Isomap',t1-t0))
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight')    

# Perform Multi-dimensional scaling
t0 = time()
trans_data = manifold.MDS(n_components=2,max_iter=100,n_init=1).fit_transform(sphere_data).T
t1 = time()
print('{}: {:.2f} sec'.format('MDS',t1-t0))
ax = fig.add_subplot(258)
plt.scatter(trans_data[0],trans_data[1],c=colors,cmap=plt.cm.rainbow)
plt.title('{} ({:.2f} sec)'.format('MDS',t1-t0))
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight')  

# Perform Spectral Embedding
t0 = time()
trans_data = manifold.SpectralEmbedding(n_neighbors=n_neighbors,n_components=2).fit_transform(sphere_data).T
t1 = time()
print('{}: {:.2f} sec'.format('Spectral Embedding',t1-t0))
ax = fig.add_subplot(259)
plt.scatter(trans_data[0],trans_data[1],c=colors,cmap=plt.cm.rainbow)
plt.title('{} ({:.2f} sec)'.format('Spectral Embedding',t1-t0))
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight')  

# Perform t-distribued stochastic neighbor embedding
t0 = time()
trans_data = manifold.TSNE(n_components=2,init='pca',random_state=2018).fit_transform(sphere_data).T
t1 = time()
print('{}: {:.2f} sec'.format('t-SNE',t1-t0))
ax = fig.add_subplot(2,5,10)
plt.scatter(trans_data[0],trans_data[1],c=colors,cmap=plt.cm.rainbow)
plt.title('{} ({:.2f} sec)'.format('t-SNE',t1-t0))
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight') 
```
